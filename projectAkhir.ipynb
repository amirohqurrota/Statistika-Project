```python
import numpy as np
import pandas as pd
from scipy.stats import binom

df=pd.read_csv('chocolate.csv', error_bad_lines=False)
compList=df["company"].values.tolist()
compListt=[]
for i in compList:
    if i not in compListt:
        compListt.append(i)
```

```python
def Probability(company):
    dfCompany=df[df["company"] == company]
    rateList=dfCompany["rating"].values.tolist()
    rateDict= {'Good' : int(), 'Bad': int()}
    probDict = {'Good' : int(), 'Bad': int()}
    for i in rateList:
        if i>=4.00:
            rateDict["Good"]+=1
        else:
            rateDict["Bad"]+=1
            
    for i in rateDict:
        probDict[i]=rateDict[i]/len(rateList)
    
    return probDict

def BinomProb(probDict):
    n = 100 #jumlah review
    p = probDict["Good"] #probabilitas diatas 4.00
    r_values = list(range(n + 1))
    dist = [binom.pmf(r, n, p) for r in r_values ]

    result=int()
    for i in range (51,101,1):
        result+=dist[i]
    return (result)

def CountSumOfRating(company):
    nameOfComp=df[df["company"] == company]
    return nameOfComp.shape[0]
```

```python
resultDict={'Company' : [], 'Probability': []} 
for i in compListt:
    i=str(i)
    a=Probability(i)
    a=BinomProb(a)
    resultDict["Company"].append(i)
    resultDict["Probability"].append(a)

# pd.set_option('display.max_rows', None)
dfBinom = pd.DataFrame.from_dict(resultDict)
# dfBinom
```

```python
# filtering dfBinom for final result

# delete company that have 0 probability
dfBinom.loc[dfBinom['Probability'] != 0]

# filter company that rated at least 20 ratings
for i in dfBinom["Company"].values.tolist():
    if CountSumOfRating(i)<20:
        dfBinom.drop(dfBinom[(dfBinom.Company == i)].index, inplace=True)
        
# sorting by Probability values
dfBinom['Probability']=dfBinom['Probability'].astype(float)
dfBinom=dfBinom.sort_values(by='Probability', ascending=False)
dfBinom
```
